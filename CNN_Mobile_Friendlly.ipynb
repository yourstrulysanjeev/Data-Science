```
import os
from keras.preprocessing import image
from keras.utils import np_utils

labels_dict = {
    'Aerodactyl':0,
    'Bulbasaur':1,
    'Charmander':2,
    'Dratini':3,
    'Fearow':4,
    'Meowth':5,
    'Pikachu':6,
    'Psyduck':7,
    'Spearow':8,
    'Squirtle':9
}
img_data = []
labels = []
```

```
for pokemon in labels_dict:
  pokemon_images = os.listdir(pokemon)
  for pokemon_img in pokemon_images:
    if '.svg' in pokemon_img:
      pass
    else:
      img_path = os.path.join(pokemon+'/',pokemon_img)
      img = image.load_img(img_path,target_size=(224,224))
      img_arr = image.img_to_array(img)
      img_data.append(img_arr)
      labels.append(labels_dict[pokemon])
```

```
import random
import numpy as np

combined = list(zip(img_data,labels))
random.shuffle(combined)
img_data[:],labels[:] = zip(*combined)
X_Train,Y_Train = np.array(img_data),np.array(labels)
Y_Train = np_utils.to_categorical(Y_Train)
X_Train.shape,Y_Train.shape
```

```
import matplotlib.pyplot as plt

for i in range(10):
  img = X_Train[i]/255.0
  plt.imshow(img)
  plt.show()
```

```
from keras.applications.mobilenet import MobileNet
from keras.applications.mobilenet_v2 import MobileNetV2
from keras.applications.resnet50 import ResNet50
from keras.layers import *
from keras.optimizers import Adam
from keras.models import Model
```

```
model = MobileNet(input_shape=(224,224,3),include_top=False,weights='imagenet')
```

```
model.summary()
```

```
av1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(256,activation='relu')(av1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(10,activation='softmax')(d1)

model_new = Model(input=model.input,outputs=fc2)
```

```
model_new.summary()
```

```
adam = Adam(learning_rate=0.00003)
model_new.compile(loss='categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
```

```
for ix in range(len(model_new.layers)):
  print(ix,model_new.layers[ix])
```

```
for ix in range(81):
  model_new.layers[ix].trainable = False

model_new.compile(loss='categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
model_new.summary()
```

```
hist = model_new.fit(X_Train,Y_Train,shuffle=True,batch_size=20,epochs=4,validation_split=0.20)
```

```
hist.history
```

```
plt.plot(hist.history['val_loss'])
plt.show()
```
