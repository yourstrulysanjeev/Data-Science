```python
f = open('glove.6B.50d.txt',encoding='utf-8')
embeddings_index = {}
```

```python
import numpy as np

for line in f:
  values = line.split()
  word = values[0]
  coefs = np.asarray(values[1:],dtype='float')
  embeddings_index[word] = coefs

f.close()
```

```python
len(embeddings_index)
```

```python
def embedding_output(X): #Accepts a 2-D matrix and returns a 3-D Tensor(Vector)
  batch_size = X.shape[0]
  max_len = 10
  embedding_dim = 50
  embedding_out = np.zeros((batch_size,max_len,embedding_dim))

  for ix in range(batch_size):
    list_of_words = X[ix].split()
    itrs = max_len

    if max_len>len(list_of_words):
      itrs = len(list_of_words)

    for jx in range(itrs):
      embedding_out[ix][jx] = embeddings_index[list_of_words[jx].lower()]
    
  return embedding_out
```

```python
embedding_matrix_Train = embedding_output(X_Train)
embedding_matrix_Test = embedding_output(X_Test)
```

```python
embedding_matrix_Train.shape,embedding_matrix_Test.shape
```

```python
# Define the RNN/LSTM Model
from keras.models import Sequential
from keras.layers import *

model = Sequential()
model.add(LSTM(64,input_shape=(10,50),return_sequences=True)) 
model.add(Dropout(0.5))
model.add(LSTM(64,return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(5))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
from keras.utils import np_utils

Y_Train_h = np_utils.to_categorical(Y_Train)
Y_Test_h = np_utils.to_categorical(Y_Test)
```

```python
from keras.callbacks import EarlyStopping,ModelCheckpoint

checkpoint = ModelCheckpoint('best_model.h5',monitor='val_acc',verbose=True,save_best_only=True)
earlystop = EarlyStopping(monitor='val_acc',patience=10)
hist = model.fit(embedding_matrix_Train,Y_Train_h,epochs=100,batch_size=64,shuffle=True,validation_split=0.20,callbacks=[earlystop,checkpoint])
```

```python
pred = model.predict_classes(embedding_matrix_Test)
```

```python
training_acc = model.evaluate(embedding_matrix_Train,Y_Train_h)
testing_acc = model.evaluate(embedding_matrix_Test,Y_Test_h)
```
